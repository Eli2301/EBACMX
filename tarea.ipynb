```python
## En este script el número de veces que se repite cada una de las palabras

simbolos = ['¿','?','.','.',';',':','¡','!']
numpalabras = 0
dpalabras = dict()
with open('elramoazul.txt', encoding='utf-8') as fichero:
    for linea in fichero:
        for simbolo in simbolos:
            linea = linea.replace(simbolo,' ')
        palabras = linea.split()
        for palabra in palabras:
            numpalabras += 1
            dpalabras[palabra] = dpalabras.get(palabra , 0) + 1
print('El texto contiene %s palabras' %numpalabras)
print('Las palabras se han repetido de la siguiente forma:')
for palabra in dpalabras:
    veces = 'veces'
    if dpalabras[palabra] == 1:
        veces = 'vez'
    print('La palabra "%s" se ha repetido %s %s' %(palabra, dpalabras[palabra], veces))
```

```python
## En este script se ordena alfabéticamente las palabras del texto

palabras=set()
with open ('elramoazul.txt', encoding='utf-8') as f:
    for palabra in f.readline().split():
        palabras.add(palabra)
 
print(sorted(list(palabras))) 
```
